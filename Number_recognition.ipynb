```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sn
```

```python
# Load the MNIST dataset into training and testing sets
(X_train, Y_train) , (X_test, Y_test) = keras.datasets.mnist.load_data()
```

```python
# Check the dimensions of the loaded dataset
print("X_train shape is : ", X_train.shape)
print("Y_train shape is : ", Y_train.shape)
print("X_test shape is : ", X_test.shape)
print("Y_test shape is : ", Y_test.shape)
```

```python
# Display an example image from the training data
plt.imshow(X_train[1010], cmap ='grey')
plt.axis("off")
```

```python
# Scale pixel values to 0-1 and reshape into 2D format for the CNN
X_train_cnn = (X_train / 255.0).reshape(-1, 28, 28, 1)
X_test_cnn = (X_test / 255.0).reshape(-1, 28, 28, 1)
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(28, 28, 1)),
    keras.layers.Conv2D(16, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_cnn, Y_train, epochs=10, validation_data=(X_test_cnn, Y_test))
```

```python
# Evaluate the trained model's performance on the unseen test data
model.evaluate(X_test_cnn, Y_test)
```

```python
# Generate predictions for the testing dataset
y_predict = model.predict(X_test_cnn)
```

```python
# Convert prediction probabilities into actual class labels
y_predict_labels = [int(np.argmax(i)) for i in y_predict]
```

```python
# Generate a confusion matrix to compare predictions against actual labels
cm = tf.math.confusion_matrix(labels = Y_test, predictions = y_predict_labels)
```

```python
# Generate a confusion matrix to compare predictions against actual labels
cm = tf.math.confusion_matrix(labels = Y_test, predictions = y_predict_labels)
```

```python
# Visualize the confusion matrix using a seaborn heatmap
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot =True, fmt='d')
plt.xlabel('predicted')
plt.ylabel('truth')
```
